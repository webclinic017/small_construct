```python
import yfinance as yf, pandas as pd
from ta import add_all_ta_features
from ta.utils import dropna

# https://technical-analysis-library-in-python.readthedocs.io/en/latest/
```

```python
tick = yf.Ticker(str('MSFT'))

# Tells yfinance what kind of data we want about this stock (In this example, all 
hist_data = tick.history(period="max")  

# Observe the historical stock data
hist_data.head() 
```

```python
hist_data = add_all_ta_features(hist_data, open="Open", high="High", low="Low", close="Close", volume="Volume") # Substantiate data with momentum indicators
hist_data.columns
```

```python
hist_data
```

```python
from ta.volatility import BollingerBands
indicator_bb = BollingerBands(close=hist_data["Close"], window=20, window_dev=2)
```

```python
# Add Bollinger Bands features
fd = pd.DataFrame()
fd['actual'] = hist_data.Close
fd['bb_bbm'] = indicator_bb.bollinger_mavg()
fd['bb_bbh'] = indicator_bb.bollinger_hband()
fd['bb_bbl'] = indicator_bb.bollinger_lband()

# Add Bollinger Band high indicator
fd['bb_bbhi'] = indicator_bb.bollinger_hband_indicator()

# Add Bollinger Band low indicator
fd['bb_bbli'] = indicator_bb.bollinger_lband_indicator()

fd.tail()
```
