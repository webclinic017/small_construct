# IMPORT SEQUENCE:

```python
start = datetime(2019, 1, 1)
end = datetime.now()

def comp_scraper(company_name):
    df = wb.DataReader(company_name, 'yahoo', start, end)
    df['comp']=company_name
    return (df)

data_df=pd.DataFrame()

company_list = tickers
tic = company_list[0]

for company in company_list:
    df=comp_scraper(company)
    data_df = data_df.append(df)

columns = df.columns
cols = [column.lower() for column in columns]
df.columns = cols

##checking shape of data frame
print(data_df.shape)
print(data_df.columns)

# Statastics
print(data_df.describe())
df.head()
```

```python
for company in company_list:
    x = data_df[data_df['comp']==company]
    x['Close'].plot(figsize=(10, 7))
    plt.title("Close Price of %s" % company, fontsize=16)
    plt.ylabel('Price', fontsize=14)
    plt.xlabel('Year', fontsize=14)
    plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
    plt.show();
```

```python
data = pd.DataFrame()
data = wb.get_data_yahoo(company_list, start = start, end = end)

Alpha = ((data['Close'].pct_change()+1).cumprod())
Alpha.plot(figsize=(13, 9))

plt.legend()
plt.title("Change (%) vs Time: " + str(company_list) + ".", fontsize=16)
plt.yticks()
plt.ylabel('Percent Change', fontsize=14)
plt.xlabel('Time', fontsize=14)

plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)
plt.show()
print(tic);
```
